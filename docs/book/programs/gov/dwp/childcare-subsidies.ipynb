## [Tax-free childcare](https://www.gov.uk/tax-free-childcare?step-by-step-nav=d78aeaf6-1747-4d72-9619-f16efb4dd89d)

(Legislation [link](https://www.legislation.gov.uk/ukdsi/2015/9780111127063))

### Financial support

The program provides up to £2,000 per year for each eligible child to support various childcare arrangements. For children with disabilities, the support is enhanced to £4,000 annually. This support is distributed quarterly, with standard payments capped at £500 per three months per child, while disabled children can receive up to £1,000 per three months.

### Operation

The tax-free childcare scheme is designed to provide financial support through a government contribution system. The key operational features include:

* For every £8 deposited by parents into their designated childcare account, the government contributes an additional £2  
* The scheme can be used simultaneously with other government childcare initiatives, specifically the 15 or 30 hours free childcare programs, provided eligibility requirements are met for both schemes  
* The household income threshold is strictly enforced \- if either partner's adjusted net income exceeds £100,000 in the current tax year, the family becomes ineligible for the program

### Eligibility

**Child age requirements**

Children must be under 12 years of age to qualify for the standard scheme. Eligibility automatically terminates on September 1st following the child's 11th birthday. For children with disabilities, the age limit is extended \- they remain eligible until September 1st following their 16th birthday, provided they receive qualifying disability benefits or are certified as blind or severely sight-impaired.

**Work status requirements**

Eligibility is closely tied to employment status. Qualifying work situations include:

1. Active employment: Applicants must be currently employed or self-employed  
2. Leave status: Individuals on the following types of leave qualify:  
   1. Sick leave  
   2. Annual leave  
   3. Shared parental leave  
   4. Maternity leave  
   5. Paternity leave  
   6. Adoption leave  
3. Return to work: Applicants must be returning to work within 31 days of their initial application date

**Non-working individual eligibility**

The scheme recognizes certain circumstances where one partner may not be working. Eligibility can be maintained if one partner is employed while the other receives any of the following benefits:

* Incapacity benefit  
* Severe disablement allowance  
* Carer's allowance (with special provision for carer support payment in Scotland)  
* Contribution-based employment and support allowance

**Minimum earnings requirements**

The scheme establishes minimum earning thresholds that must be met by each partner over a three-month period:

* Ages 21 and over: Minimum earnings of £2,379  
* Ages 18-20: Minimum earnings of £1,788  
* Under 18 or apprentices: Minimum earnings of £1,331

These thresholds are based on national minimum wage or living wage calculations for 16 hours per week on average.

### Key considerations

**Income exclusions**

Certain types of income are specifically excluded from eligibility calculations:

* Dividend payments  
* Interest earned  
* Income from property investments  
* Pension payments

**Special provisions for self-employed**

Newly established self-employed individuals (business less than 12 months old) benefit from more flexible earnings requirements. They may qualify with lower earnings during their initial business period.

**Income calculations**

The scheme offers flexibility in how income is calculated:

* Multiple income sources can be combined to meet the minimum thresholds  
* Annual income averaging is permitted (e.g., irregular earnings totaling £9,518 annually can qualify by averaging to £2,379 quarterly)  
* Self-employed individuals can choose to use either their total income or solely their self-employment income if it provides eligibility

### Program incompatibilities

**Mutually exclusive benefits**

The tax-free childcare scheme cannot be combined with:

* Working tax credit  
* Child tax credit  
* Universal credit  
* Childcare vouchers or directly contracted childcare

**Impact on existing benefits**

Upon successful application:

* Any existing tax credits will be immediately terminated  
* Reapplication for terminated tax credits is not permitted  
* Childcare vouchers or directly contracted childcare must be cancelled within 90 days  
* Universal credit claims should only be cancelled after receiving confirmation of tax-free childcare approval  
* Childcare bursaries or grants cannot be received simultaneously with tax-free childcare

## [Free childcare](https://www.gov.uk/check-eligible-free-childcare-if-youre-working?step-by-step-nav=f517cd57-3c18-4bb9-aa8b-1b907e279bf9)

(Legislation [link](https://www.legislation.gov.uk/uksi/2017/1160/pdfs/uksiem_20171160_en.pdf))

**Age-based entitlement**

The program provides free childcare hours based on the child's age:

* Children aged 9 months to 2 years are eligible for 15 hours of free childcare per week  
* Children aged 3 to 4 years are eligible for 30 hours of free childcare per week

### Employment eligibility criteria

Eligibility is based on the employment status of primary caregivers and their partners. Qualifying situations include:

* Current employment or about to start new employment  
* Individuals on various types of leave:  
  * Sick leave  
  * Annual leave  
  * Shared parental leave  
  * Maternity leave  
  * Paternity leave  
  * Adoption leave

### Term commencement dates

For those starting new employment, free childcare availability aligns with term dates:

* Employment starting between May and September: Term begins on or after 1 September  
* Employment starting between October and January: Term begins on or after 1 January  
* Employment starting between February and April: Term begins on or after 1 April

### Alternative eligibility conditions

Eligibility may be maintained if one partner is working while the other receives any of the following benefits:

* Incapacity benefit  
* Severe disablement allowance  
* Carer's allowance  
* Limited capability for work benefit  
* Employment and support allowance (contribution-based)

### Income requirements

Minimum earnings thresholds must be met on a quarterly basis for each individual:

* Ages 21 and over: £2,380 per quarter (equivalent to £183 weekly)  
* Ages 18-20: £1,788 per quarter (equivalent to £137 weekly)  
* Under 18 or apprentices: £1,331 per quarter (equivalent to £102 weekly)

These thresholds are based on national minimum wage or living wage calculations for 16 hours per week on average.

### Income assessment methods

The scheme provides flexible options for income assessment:

* Annual income averaging is permitted for those with irregular earnings patterns  
* Multiple income sources can be combined to meet the minimum thresholds  
* Self-employed individuals can choose to use their self-employment income independently if this provides eligibility  
* Newly established self-employed individuals (business less than 12 months old) may qualify with lower earnings

### Income exclusions

Certain types of income are specifically excluded from eligibility calculations:

* Dividend payments  
* Interest earned  
* Income from property investments  
* Pension payments

### Additional considerations

**Income limit**

Individuals or partners with an expected adjusted net income over £100,000 in the current tax year are not eligible

**Foster carers**

Foster carers may be eligible if they:

* Are in paid work outside their fostering role  
* Have an adjusted net income under £100,000  
* Must apply through their social worker and local authority rather than online

## [15 hours free childcare](https://www.gov.uk/help-with-childcare-costs/free-childcare-and-education-for-3-to-4-year-olds?step-by-step-nav=f237ec8e-e82c-4ffa-8fba-2a88a739783b)

(Legislation [link](https://researchbriefings.files.parliament.uk/documents/CBP-8054/CBP-8054.pdf.#:~:text=All%20three%20and%20four%2Dyear%2Dolds%20in%20England%20are%20eligible,and%20applies%20irrespective%20of%20income.) (Childcare Act 2006 \- page 10\))

### Core entitlement

The scheme provides universal free childcare for all 3-4 year olds in England:

* All eligible children receive 570 hours of free childcare annually  
* The standard delivery format is 15 hours per week across 38 weeks of the year  
* Parents may choose to distribute hours differently if providers offer flexibility  
* Some children may qualify for expansion to 30 hours scheme based on additional eligibility criteria

### Implementation requirements

**Timing**

Free childcare provision begins from the term following the child's third birthday, with specific start dates:

* 1 January  
* 1 April  
* 1 September

**Provider requirements**

The free early education and childcare must be:

* Delivered through an approved childcare provider  
* Continued until the child starts reception class or reaches compulsory school age, whichever is later

### Additional costs

Parents may need to cover expenses beyond the free entitlement:

* Meal costs during childcare hours  
* Nappy supplies and changes  
* Any hours exceeding the free entitlement allocation  
* Additional activities and programs 

### Regional variations

The scheme as described applies to England only. Different programs operate in: Scotland, Wales, and Northern Ireland. Each nation maintains its own distinct childcare support schemes and system

## [Free education and childcare for 2-year-olds](https://www.gov.uk/help-with-childcare-costs/free-childcare-2-year-olds-claim-benefits?step-by-step-nav=f237ec8e-e82c-4ffa-8fba-2a88a739783b)

(Legislation [link](https://www.legislation.gov.uk/uksi/2020/712/pdfs/uksiem_20200712_en.pdf))

### Benefit-based qualification

Eligibility is granted to families receiving any of the following benefits:

* Income support  
* Income-based jobseeker's allowance (JSA)  
* Income-related employment and support allowance (ESA)  
* Universal credit with household income of £15,400 or less annually after tax, excluding benefit payments  
* Pension credit (guaranteed element)  
* Child tax credit and/or working tax credit with household income of £16,190 or less annually before tax  
* Working tax credit 4-week run on period (payment received when ceasing to qualify for working tax credit)

### Alternative eligibility criteria

Children may qualify if they meet any of the following conditions:

* Are under local authority care  
* Have an education, health and care (EHC) plan  
* Receive disability living allowance

### Immigration status considerations

For non-UK residents with 'no recourse to public funds' status, eligibility depends on household income thresholds after tax:

**Outside London**

* Single child families: £26,500  
* Multiple children families: £30,600

**Within London**

* Single child families: £34,500  
* Multiple children families: £38,600

**Additional requirement**

* Total savings and investments must not exceed £16,000

**Additional costs**

Parents may need to cover expenses beyond the free entitlement:

* Meal costs during childcare hours  
* Nappy supplies and changes  
* Hours exceeding the free entitlement allocation

### Program restrictions

Important limitations include:

* The scheme cannot be combined with the working parents' childcare scheme  
* If eligible for both schemes, applications must be made through the benefits-based program  
* Eligibility must be verified through local council or childcare provider

### Regional variations

The scheme as described applies to England only. Different programs operate in: Scotland, Wales, and Northern Ireland. Each nation maintains its own distinct childcare support schemes and systems.
